```python
columns_name = ['Intermittency', 'Number of metastases', 'Status',
       '1Synchron/2Metachron', 'Nephrectomy', 'MSKCC', 'Differentiation',
       'Ray coding', 'Nexavar', 'Metastatic surgery', 'Variant',
       'Sex 1 - m 2 - w', 'Progress code',
       'ВПП']
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.ensemble import GradientBoostingClassifier
```

```python
def code(x):
    if x <= 24:      #До 2 лет
        return 0
    return 1   
```

```python
df['ВПП_code'] = df['ВПП'].apply(code)
```

```python
df
```

```python
from sklearn. model_selection import train_test_split
```

```python
train_data = pd.DataFrame()
test_data = pd.DataFrame()
for i in range(0, 2, 1):
    a = df[df['ВПП_code'] == i]
    train, test = train_test_split(a, test_size=0.2, random_state=3)
    train_data = pd.concat([train_data, train], ignore_index=True)
    test_data = pd.concat([test_data, test], ignore_index=True)

train_data = train_data.sample(frac=1)  #перемешиваю
test_data = test_data.sample(frac=1)
```

```python
columns_name = ['Intermittency', 'Number of metastases', 'Status',
       '1Synchron/2Metachron', 'Nephrectomy', 'MSKCC', 'Differentiation',
       'Ray coding', 'Nexavar', 'Metastatic surgery', 'Variant',
       'Sex 1 - m 2 - w', 'Progress code']
```

```python
X_tr = train_data[columns_name]
y_tr = train_data['ВПП']

X_ts = test_data[columns_name]
y_ts = test_data['ВПП']
```

```python
from sklearn.model_selection import GridSearchCV
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import fbeta_score, make_scorer, classification_report, roc_curve, confusion_matrix, roc_auc_score, accuracy_score, f1_score

#Gradient Boost
param_grid = {'n_estimators': [5, 10],
              'learning_rate': [0.1, 0.04, 0.05, 0.02],
              'max_depth': [4],
              'min_samples_leaf': [3],
              'max_features': [1.0]}
n_jobs = 4

estimator = GradientBoostingClassifier(random_state=3)
gb_classifier = GridSearchCV(estimator=estimator,
                             cv=5,
                             param_grid=param_grid,
                             n_jobs=n_jobs)

best_gb_model = gb_classifier.fit(X_tr, train_data['ВПП_code'])
print('Наилучшие параметры для Gradient Descent: {}'.format(best_gb_model.best_params_))
gb_classifier = gb_classifier.best_estimator_
gb_classifier.get_params()

estimator = gb_classifier.fit(X=X_tr, y=train_data['ВПП_code'])
#Predict
y_pred_gb = estimator.predict(X_ts)
print('Точность измерений (тренировочные данные): ',
      accuracy_score(y_true=train_data['ВПП_code'], y_pred=estimator.predict(X_tr)))
print('Точность измерений (тестовые данные): ', accuracy_score(y_true=test_data['ВПП_code'], y_pred=y_pred_gb))

report = classification_report(test_data['ВПП_code'], estimator.predict(X_ts))
print(report)

sns.set(font_scale=1.5)
sns.set_color_codes("muted")

plt.figure(figsize=(10, 8))
fpr, tpr, thresholds = roc_curve(test_data['ВПП_code'], estimator.predict(X_ts), pos_label=1)
lw = 2
plt.plot(fpr, tpr, lw=lw, label='ROC curve ')
plt.plot([0, 1], [0, 1])
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.savefig("ROC.png")
plt.show()

```

```python
import joblib

# Сохранение модели после обучения
joblib.dump(estimator, 'gb_classifier_model.pkl')
print("Модель сохранена как 'gb_classifier_model.pkl'")
```

```python
import pandas as pd
from lifelines import CoxPHFitter


# Преобразуем столбец '1Жив_ 2умер' в бинарный формат (1 = событие не наступило, 2 = событие наступило)
df['event'] = df['1Жив_ 2умер'].apply(lambda x: 0 if x == 1 else 1)

# Обучающие данные для модели
df['duration'] = df['ВПП']

# Выбираем столбцы для обучения модели (не учитываем целевые переменные '1Жив_ 2умер' и 'ВПП')
features = [
    'Intermittency', 'Number of metastases', 'Status', '1Synchron/2Metachron', 'Nephrectomy',
    'MSKCC', 'Differentiation', 'Ray coding', 'Nexavar', 'Metastatic surgery', 'Variant',
    'Sex 1 - m 2 - w', 'Progress code'
]

# Убедимся, что в данных нет пропущенных значений
df = df[features + ['duration', 'event']].dropna()

# Обучение модели Кокса
cph = CoxPHFitter()
cph.fit(df, duration_col='duration', event_col='event')
```

```python
# Оценка модели

# 1. Логарифм правдоподобия
print(f"Log-Likelihood: {cph.log_likelihood_}")

# 2. C-Index (Конкорданс Индекс)
c_index = cph.concordance_index_
print(f"C-Index: {c_index:.3f}")

# 3. Проверка предположения о пропорциональных рисках
cph.check_assumptions(df, p_value_threshold=0.05)  # Проверим на уровне 0.05

# 4. Остаточные данные
cph.plot()
```

```python
import pickle

joblib.dump(cph, 'cox_model.pkl')

print("Модель сохранена!")
```
